```python
import numpy as np
import torchvision
from matplotlib import pyplot as plt
```

```python
from NAT_extended.data_providers.aircraft import FGVCAircraftDataProvider
from NAT_extended.data_providers.blood_cells import BloodCellsDataProvider
```

```python
def imshow(img):
    img = img / 2 + 0.5
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()
```

```python
def show_imgs(dataiter):
    for _ in range(1):
        # get some random training images
        images, labels = dataiter.next()
        # show images
        print(labels[:2])
        imshow(torchvision.utils.make_grid(images[:2]))
```

```python
def print_train(provider):
    print("TRAINING\n")
    dataiter = iter(provider.train)
    show_imgs(dataiter)

def print_valid(provider):
    print("VALIDATION\n")
    dataiter = iter(provider.valid)
    show_imgs(dataiter)

def print_test(provider):
    print("TESTING\n")
    dataiter = iter(provider.test)
    show_imgs(dataiter)
```

```python
def print_all(provider):

    print(provider.__dict__)
    print("\n")

    print(len(provider.train.dataset))
    print(len(provider.valid.dataset))
    print(len(provider.test.dataset))
    print(len(provider.train))
    print(len(provider.valid))
    print(len(provider.test))
    print("\n")

    print(provider.build_train_transform())
    print("\n")

    print(provider.build_valid_transform())
    print("\n")
```
